```python
# importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import pandas_datareader as web
import datetime
from datetime import date

%matplotlib inline
```

```python
# Fetching Nike data from Yahoo! Finance
Nike = web.get_data_yahoo('NKE',start = datetime.datetime(2018, 1, 2), 
                             end = date.today())
```

```python
Nike = Nike.drop(columns=['Volume', 'Adj Close'])
Nike.head()
```

```python
# Plotting Nike Close Price over the years
fig = go.Figure()
fig.add_trace(go.Scatter(x= Nike.index, y= Nike.Close, name='Close'))
fig.update_layout(title='Price of Nike over the Years', xaxis_title='Date', yaxis_title='Price')
fig.show()
```

# Simple Moving Average (SMA)

```python
# Fast SMA (50 days)
Nike['SMA_50'] = Nike['Close'].rolling(50).mean()

# Slow SMA (200 days)
Nike['SMA_200'] = Nike['Close'].rolling(200).mean()

Nike = Nike.dropna()
```

```python
Nike.head()
```

```python
# Plotting daily close price, SMA_50 and SMA_200

fig_SMA = go.Figure()
fig_SMA.add_trace(go.Scatter(x= Nike.index, y= Nike.Close, name='Close'))
fig_SMA.add_trace(go.Scatter(x= Nike.index, y= Nike.SMA_50, name='SMA_50'))
fig_SMA.add_trace(go.Scatter(x= Nike.index, y= Nike.SMA_200, name='SMA_200'))

fig_SMA.update_layout(title='Price History of Nike and SMA',
                   xaxis_title='Date',
                   yaxis_title='Price')
fig_SMA.show()
```

# Generating Trade Signals using SMA

```python
# 'Buy' signal if Short SMA > Long SMA else 'Sell'
Nike['Signal_SMA'] = np.where(Nike['SMA_50'] > Nike['SMA_200'], 1.0, 0.0)
Nike['Position_SMA'] = Nike['Signal_SMA'].diff()
Nike.head(250)
```

```python
fig_SMA
```

```python
fig_SMA.add_trace(go.Scatter(
                         mode= 'markers',
                         name='Buy',
                         x= Nike[Nike['Position_SMA']==1].index,
                         y = Nike[Nike['Position_SMA']==1].SMA_200,
                         marker=dict(
                                     color='green',
                                     size=15,
                                     symbol='triangle-up')
                        
                        )
             )

fig_SMA.add_trace(go.Scatter(
                         mode= 'markers',
                         name='Sell',
                         x= Nike[Nike['Position_SMA']==-1].index,
                         y = Nike[Nike['Position_SMA']==-1].SMA_200,
                         marker=dict(
                                     color='red',
                                     size=15,
                                     symbol='triangle-down'
                                     )
                        
                        )
             )
```

# Back-testing SMA Strategy

```python
Nike['Position_SMA'].value_counts()
```

```python
Nike.columns.get_loc('Position_SMA')
```

```python
Nike.columns.get_loc('Close')
```

```python
Nike.shape[0]
```

```python
buyAmt = 0
sellAmt = 0
buyDates = np.array([])

for i in range(Nike.shape[0]):
    if Nike.iloc[i, 7]==1:
        buyAmt = buyAmt + Nike.iloc[i, 3]*100
        buyDates = np.append(buyDates, i)

print('Total Amount Invested:',buyAmt)

for i in range(Nike.shape[0]):
    for j in buyDates:
        if i == j:
            if (int(j)+60 < Nike.shape[0]):
                sellAmt = sellAmt + Nike.iloc[int(j+60), 3]*100
            else:
                sellAmt = sellAmt + Nike.iloc[int(j+10), 3]*100

total_SMA = sellAmt - buyAmt

print('Total Cumulative Returns:',total_SMA)
print('Returns in %:',(sellAmt/buyAmt)*100)
```

# Exponential Moving Average (EMA)

```python
# Fast EMA (50 days)
Nike['EMA_50'] = Nike['Close'].ewm(span= 50, adjust=False).mean()

# Slow EMA (200 days)
Nike['EMA_200'] = Nike['Close'].ewm(span= 200, adjust=False).mean()

Nike.head()
```

# Generating Trade Signals using EMA

```python
# 'Buy' signal if Short EMA > Long EMA else 'Sell'
Nike['Signal_EMA'] = np.where(Nike['EMA_50'] > Nike['EMA_200'], 1.0, 0.0)
Nike['Position_EMA'] = Nike['Signal_EMA'].diff()
Nike.head()
```

```python
# Plotting daily close price, EMA_50, EMA_200 and generating trade signals

fig_EMA = go.Figure()
fig_EMA.add_trace(go.Scatter(x= Nike.index, y= Nike.Close, name='Close'))
fig_EMA.add_trace(go.Scatter(x= Nike.index, y= Nike.EMA_50, name='EMA_50'))
fig_EMA.add_trace(go.Scatter(x= Nike.index, y= Nike.EMA_200, name='EMA_200'))

fig_EMA.update_layout(title='Price History of Nike and EMA',
                   xaxis_title='Date',
                   yaxis_title='Price')

fig_EMA.add_trace(go.Scatter(
                         mode= 'markers',
                         name='Buy',
                         x= Nike[Nike['Position_EMA']==1].index,
                         y = Nike[Nike['Position_EMA']==1].EMA_200,
                         marker=dict(
                                     color='green',
                                     size=15,
                                     symbol='triangle-up')
                        
                        )
             )

fig_EMA.add_trace(go.Scatter(
                         mode= 'markers',
                         name='Sell',
                         x= Nike[Nike['Position_EMA']==-1].index,
                         y = Nike[Nike['Position_EMA']==-1].EMA_200,
                         marker=dict(
                                     color='red',
                                     size=15,
                                     symbol='triangle-down'
                                     )
                        
                        )
             )

fig_EMA.show()
```
